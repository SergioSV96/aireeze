```python
import pandas as pd
```

```python
weather_stations = pd.read_json('../data/raw/weather_stations.json')
weather_stations
```

```python
# Change column order
weather_stations = weather_stations[['indicativo',
 'nombre',
 'provincia',
 'indsinop',
 'latitud',
 'longitud',
 'altitud'
 ]]
```

```python
weather_stations = weather_stations.query("provincia == 'MADRID'")
```

```python
def dms_to_dd(dms):
    # converting geographical coordinates system from DMS format to DD format
    # dms input format must be a tuple (degrees, minutes, seconds, N/S/E/W)
    directions = {
        'N': 1,
        'S': -1,
        'E': 1,
        'W': -1
    }
    return (dms[0] + dms[1]/60 + dms[2]/3600) * directions[dms[3]]


dms = "404735N"
dms_tuple = (int(dms[:2]), int(dms[2:4]), int(dms[4:6]), dms[-1])
print(dms_tuple)
```

```python
# Change geographical coordinates system
weather_stations["latitud"] = weather_stations["latitud"].apply(lambda x: dms_to_dd((int(x[:2]), int(x[2:4]), int(x[4:6]), x[-1])))
weather_stations["longitud"] = weather_stations["longitud"].apply(lambda x: dms_to_dd((int(x[:2]), int(x[2:4]), int(x[4:6]), x[-1])))
weather_stations
```

```python
# Export the processed data
weather_stations.to_csv('../data/processed/weather_stations.csv', index=False)
```
